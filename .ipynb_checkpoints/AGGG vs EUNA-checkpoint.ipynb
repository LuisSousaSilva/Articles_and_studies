```python
Tickers = ['IWDA.AS']
ETF_Tickers =['IWDA']

ETFs2 = pd.DataFrame()

# Download
for ticker in Tickers:
    try:
        url = "https://eodhistoricaldata.com/api/eod/" + str(ticker) + "?api_token=5c982bff80deb2.22427269&period=d."
        ETF = pd.read_csv(url, index_col = 'Date', parse_dates = True)[['Adjusted_close']].iloc[:-1, :]
        ETFs2 = ETFs2.merge(ETF, left_index = True, right_index = True, how='outer')
    except:
        print('Download of fund ' + ticker + ' failed')
        
ETFs2.columns = ETF_Tickers
ETFs2.index = pd.to_datetime(ETFs2.index)
```

```python
fundos = pl.merge_time_series(fundos, ETFs2)
returns = fundos.pct_change().dropna()
returns.corr()
```

```python
from pandas.tseries.offsets import DateOffset
plus_1_year = returns.iloc[0].name + DateOffset(years=1)

returns['AGGG'].rolling("365D").corr(returns['EUNA'])[plus_1_year:].iplot(dimensions=pl.dimensions, color='dodgerblue', title="Correlao anual rolante entre o AGGG e o EUNA")
```

```python
returns['AGGG'].rolling("365D").corr(returns['IWDA'])[plus_1_year:].iplot(dimensions=pl.dimensions, color='dimgrey', title="Correlao anual rolante entre o AGGG e o IWDA")
```

```python
returns['EUNA'].rolling("365D").corr(returns['IWDA'])[plus_1_year:].iplot(dimensions=pl.dimensions, color='goldenrod', title="Correlao anual rolante entre o EUNA e o IWDA")
```
