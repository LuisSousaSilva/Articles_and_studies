```python
# importing libraries
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py
import cufflinks as cf
import seaborn as sns
import pandas as pd
import numpy as np
import quandl
import plotly
import time

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from IPython.display import Markdown, display
from matplotlib.ticker import FuncFormatter
from pandas.core.base import PandasObject
from datetime import datetime

# Setting pandas dataframe display options
pd.set_option("display.max_rows", 20)
pd.set_option('display.width', 800)
pd.set_option('max_colwidth', 800)

pd.options.display.float_format = '{:,.4f}'.format

# Set plotly offline
init_notebook_mode(connected=True)

# Set matplotlib style
plt.style.use('seaborn')

# Set cufflinks offline
cf.go_offline()

# Defining today's Date
from datetime import date
today = date.today()

import PortfolioLab as pl

def compute_portfolio_with_contributions(df, weights, start_value=100):
    start = df.index[0]
    Contribuição_diária_individual_all = pd.DataFrame()

    # Start and End date for backtest
    start = df.index[0]
    end = df.index[-1]

    # Calculate the rebalancing days dates (including inception date)
    start_index = pd.date_range(start=start, end=start, periods=1)
    end_index = pd.date_range(start=end, end=end, periods=1)    
    rebalancing_days = pd.date_range(start=start, end=end, freq='A').union(start_index)
    
    if end_index not in rebalancing_days:
        rebalancing_days = rebalancing_days.union(end_index)

    weights = weights
    n_assets = len(weights)
    start_value = start_value
    active_weights = weights.copy()


    for i in range(1, len(rebalancing_days)):

        # Select period between rebalances
        quotes = df[rebalancing_days[i-1]:rebalancing_days[i]]

        n_days = len(quotes)

        Evolução = quotes / quotes.iloc[0]
        
        UPs = np.tile(active_weights, n_days)        
        UPs.shape = (n_days, n_assets)
        UPs = pd.DataFrame(UPs)
        UPs.columns = quotes.columns
        UPs = UPs * 100    
        UPs.set_index(quotes.index, inplace=True)

        Evolução_no_portfolio = Evolução.mul(UPs, axis=1)

        Contribuição_do_dia = Evolução_no_portfolio - Evolução_no_portfolio.shift(1)

        Contribuição_diária_individual_all = pd.concat([Contribuição_diária_individual_all, Contribuição_do_dia.dropna()])

        Contribuição_no_portfolio = Contribuição_do_dia.cumsum()

        Portfolio = pd.DataFrame(Contribuição_diária_individual_all.dropna().cumsum().sum(axis=1)) + 100

        active_weights = (weights * np.repeat(Portfolio.iloc[-1][0], len(active_weights))) / 100
        
    Contribuição_individual = Contribuição_diária_individual_all.cumsum()    
    Contribuição_individual = Contribuição_individual[Contribuição_individual.index.dayofweek < 5]
    
    Portfolio.columns = ['Portfolio']
    Portfolio = Portfolio[Portfolio.index.dayofweek < 5]
    
    # Add initial day to portfolio   
    inicial_day_portfolio = pd.DataFrame({'Portfolio':[100]}, index=[start])
    Portfolio = pd.concat([inicial_day_portfolio, Portfolio])
    
    # Add initial day to Contribuição_individual
    ic_array = np.repeat(0, len(df.columns))
    inicial_day_ci = pd.DataFrame([ic_array], index=[start])
    inicial_day_ci.columns=df.columns
    Contribuição_individual = pd.concat([inicial_day_ci, Contribuição_individual])
    
    return Portfolio, Contribuição_individual
```

```python
pl.ichart(round(pl.compute_drawdowns(Portfolios), 2),
         yticksuffix='%',
         title='Drawdowns (de 2014-02-27 a 2021-01-07)',
         yTitle='Percentagem de queda desde os máximos')
```

Drawdown é quanto a carteira caiu desde o topo. Quanto alguém poderia perder se tivesse comprado no pico mais alto, mesmo antes de uma queda do portfolio. Naturalmente se a carteira fez novo máximo o drawdown é 0%. Neste caso podemos ver que as carteiras, face aos máximos, chegaram a cair cerca de 10% em 2015, chegoram a uma queda de 10%+ em inícios de 2016. Por outro lado a queda no início de 2020 devido ao Covid-19 foi mais profunda mas a recuperação foi mais rápida (a carteira de ETFs já fez máximo enquanto a de fundos está praticamente em máximos).

- **CAGR**: Retorno Anualizado
- **Return**: Retorno Total
- **StdDev**: Desvio Padrão (Quanto mais alto mais volátil e arriscada é uma carteira)
- **Sharpe**: CAGR a dividir por StdDev (Quanto mais alto melhor, pois o portfolio teve maior retorno por unidade de risco/StdDev)
- **Max DD**: Maior queda em termos de profundidade (o valor mínimo do gráfico dde "Drawdowns do Portfolio")
- **MAR**: CAGR a dividir por Max DD (Quanto mais alto melhor, pois o portfolio teve maior retorno por unidade de risco/Drawdown)

Em todos os períodos a maior rentabilidade foi atingida com menor risco nas carteiras de ETFs (como medido pelo desvio padrão). Portanto mesmo que a rentabilidade fosse igual as carteiras de ETFs mostravam-se superior devido ao melhor rácio retorno/risco.

A adição do ouro tem como objectivo diversificar e aumentar a possibilidade de manter rentabilidades aceitávelmente boas, embora seja expectável que não se continue a ter estas rentabilidade no futuro. Isso deve-se, entre outras coisas, à contínuada queda das taxas de juro (para a pessoa comum isso é visível pelas taxas baixíssimas dos depósitos a prazo e da euribor).

Essa questão da diversificação é notória na numa altura que a 7 anos (+/-) a carteira de ETFs sem ouro  teve uma performance ligeiramente melhor mas a 3 e a 1 ano a carteira com ouro tem vindo a ter uma melhor performance, o que nos demonstra que é mais resiliente a manter as rentabilidades. 

Todos os 3 ETFs contribuiram positivamente para a performance do portfolio, sendo que é possível ver em baixo quanto cada um dos ETFs contribuiu para a performance total.
