# Carteira 2 Fundos (análise até 2020-12-31)

Análise de carteira com apenas dois ETFs/fundos. Um de obrigações global que pode ser o **IE00B18GC888** (Vanguard Global Bd Idx EUR H Acc) ou o **VAGF** (Vanguard Global Aggt Bd ETF EUR H Acc | VAGF) e o **IWDA** (iShares Core MSCI World ETF USD Acc (EUR) | IWDA).

As carteiras têm o peso indicado nas respectivas legendas sendo o VAGF a parte obrigacionista e o IWDA a parte accionista.

```python
Quotes = pl.normalize(pl.get_quotes(['IWDA', 'IE00B18GC888'], names=['IWDA', 'VG GB']))[:'2020-12-31']
```

```python
pl.ichart(round(Quotes,2), colors=pl.colors_list, title='Performance de cada activo individualmente', yticksuffix=' €', yTitle='Valorização por cada 100 €uros investidos', colab=False)
```

```python
portfolio_25_75, contributions_25_75 = pl.compute_portfolio_with_contributions(Quotes, [0.25, 0.75])
portfolio_50_50, contributions_50_50 = pl.compute_portfolio_with_contributions(Quotes, [0.50, 0.50])
portfolio_55_45, contributions_55_45 = pl.compute_portfolio_with_contributions(Quotes, [0.55, 0.45])
```

```python
portfolios = pl.merge_time_series(portfolio_25_75, portfolio_50_50)
portfolios = pl.merge_time_series(portfolios, portfolio_55_45)
portfolios.columns = ['75O/25A', '50O/50A', '45O/55A']
```

```python
pl.ichart(round(portfolios, 2), colors=pl.colors_list, title='Performance de cada activo individualmente', yticksuffix=' €', yTitle='Valorização por cada 100 €uros investidos', colab=False)
```

```python
dd = pl.compute_drawdowns(portfolios)
pl.ichart(round(dd,2), colors =pl.colors_list, title='Drawdowns dos portfolios', yticksuffix='%', colab=False)
```
