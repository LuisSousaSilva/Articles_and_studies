```python
%matplotlib inline

# importing libraries
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py
import cufflinks as cf
import seaborn as sns
import pandas as pd
import numpy as np
import quandl
import plotly
import time

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from IPython.display import Markdown, display
from matplotlib.ticker import FuncFormatter
from pandas.core.base import PandasObject
from datetime import datetime

# Setting pandas dataframe display options
pd.set_option("display.max_rows", 50)
pd.set_option('display.width', 800)
pd.set_option('max_colwidth', 800)

# Set plotly offline
init_notebook_mode(connected=True)

# Set matplotlib style
plt.style.use('seaborn')

# Set cufflinks offline
cf.go_offline()

# Defining today's Date
from datetime import date
today = date.today()

import PortfolioAnalyser as pa
```

```python
%%time
starting_capital = 10000
savings = 250
start = '1999-01-01'
end   = '2025-01-01'

# Order logic
# List of days of orders
days_of_orders_list = pd.date_range(start=start, end=end, freq='BA')

MSCI = pd.read_excel('MSCI_USD.xls').iloc[6:].dropna()

MSCI.columns = ['Date', 'MSCI']

# Convert the date column to datetime64
MSCI.Date = pd.to_datetime(MSCI.Date)

# Set date column as index
MSCI.set_index('Date', inplace=True)

# Set Period
MSCI = pa.normalize(MSCI['1999-12':])

# Savings
MSCI['Savings'] = savings
MSCI['Savings'].iloc[0] = 0
MSCI['Savings'] = MSCI['Savings'].cumsum()

# UPs
MSCI['UPs'] = starting_capital / MSCI['MSCI'].iloc[0]

# # Buy or hold
MSCI['Buy'] = MSCI.index.isin(days_of_orders_list)
        
# Invested
MSCI['S_Invested'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['S_Invested'].iloc[i] = MSCI['Savings'].iloc[i]
    else:
        MSCI['S_Invested'].iloc[i] = None
        
MSCI['S_Invested'] = MSCI['S_Invested'].ffill()

# Savings left
MSCI['S_left'] = 0

for i in range(MSCI.shape[0]):
    MSCI['S_left'].iloc[i] = MSCI['Savings'].iloc[i] - MSCI['S_Invested'].iloc[i]

for i in range(MSCI.shape[0]):
    if MSCI['S_left'].iloc[i] == 0:
        MSCI['S_left'].iloc[i] = MSCI['S_left'].iloc[i - 1] + savings
        
MSCI['S_left'].iloc[0] = 0

# Savings in UPs
MSCI['S_UPs'] = MSCI['S_left'] / MSCI['MSCI']

# Order sizing
MSCI['order'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['order'].iloc[i] = MSCI['S_UPs'].iloc[i]
    else:
        MSCI['order'].iloc[i] = None
        
MSCI['order'] = MSCI['order'].cumsum().ffill()

MSCI['UPs_new'] = MSCI['order'] + MSCI['UPs']

MSCI['Montante'] = MSCI['UPs_new'] * MSCI['MSCI']

MSCI.iplot(dimensions=pa.dimensions)
```

```python
%%time
starting_capital = 10000
savings = 220 
start = '1999-01-01'
end   = '2025-01-01'

# Order logic
# List of days of orders
days_of_orders_list = pd.date_range(start=start, end=end, freq='BM')

MSCI = pd.read_excel('MSCI_USD.xls').iloc[6:].dropna()

MSCI.columns = ['Date', 'MSCI']

# Convert the date column to datetime64
MSCI.Date = pd.to_datetime(MSCI.Date)

# Set date column as index
MSCI.set_index('Date', inplace=True)

# Set Period
MSCI = pa.normalize(MSCI['2016-09-01':])

# Savings
MSCI['Savings'] = savings
MSCI['Savings'].iloc[0] = 0
MSCI['Savings'] = MSCI['Savings'].cumsum()

# UPs
MSCI['UPs'] = starting_capital / MSCI['MSCI'].iloc[0]

# # Buy or hold
MSCI['Buy'] = MSCI.index.isin(days_of_orders_list)
        
# Invested
MSCI['S_Invested'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['S_Invested'].iloc[i] = MSCI['Savings'].iloc[i]
    else:
        MSCI['S_Invested'].iloc[i] = None
        
MSCI['S_Invested'] = MSCI['S_Invested'].ffill()

MSCI['S_Invested'] = MSCI['S_Invested'] .shift()

# Savings left
MSCI['S_left'] = 0

for i in range(MSCI.shape[0]):
    MSCI['S_left'].iloc[i] = MSCI['Savings'].iloc[i] - MSCI['S_Invested'].iloc[i]
        
MSCI['S_Invested'].iloc[0] = 0
MSCI['S_left'].iloc[0] = 0

# Savings in UPs
MSCI['S_UPs'] = MSCI['S_left'] / MSCI['MSCI']

# Order sizing
MSCI['order'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['order'].iloc[i] = MSCI['S_UPs'].iloc[i]
    else:
        MSCI['order'].iloc[i] = None
        
MSCI['order'] = MSCI['order'].cumsum().ffill()

MSCI['UPs_new'] = MSCI['order'] + MSCI['UPs']

MSCI['Montante'] = MSCI['UPs_new'] * MSCI['MSCI']

MSCI.iplot(dimensions=pa.dimensions)
```

```python
%%time
starting_capital = 10000
savings = 10
start = '1999-01-01'
end   = '2025-01-01'

# Order logic
# List of days of orders
days_of_orders_list = pd.date_range(start=start, end=end, freq='BM')

MSCI = pd.read_excel('MSCI_USD_daily.xls').iloc[6:].dropna()

MSCI.columns = ['Date', 'MSCI']

# Convert the date column to datetime64
MSCI.Date = pd.to_datetime(MSCI.Date)

# Set date column as index
MSCI.set_index('Date', inplace=True)

# Set Period
MSCI = pa.normalize(MSCI['2016-09-30':])

# Savings
MSCI['Savings'] = savings
MSCI['Savings'].iloc[0] = 0
MSCI['Savings'] = MSCI['Savings'].cumsum()

# UPs
MSCI['UPs'] = starting_capital / MSCI['MSCI'].iloc[0]

# # Buy or hold
MSCI['Buy'] = MSCI.index.isin(days_of_orders_list)
        
# Invested
MSCI['S_Invested'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['S_Invested'].iloc[i] = MSCI['Savings'].iloc[i]
    else:
        MSCI['S_Invested'].iloc[i] = None
        
MSCI['S_Invested'] = MSCI['S_Invested'].ffill()

MSCI['S_Invested'] = MSCI['S_Invested'].shift()

MSCI['S_Invested'] = MSCI['S_Invested'].fillna(0)

# Savings left
MSCI['S_left'] = 0

for i in range(MSCI.shape[0]):
    MSCI['S_left'].iloc[i] = MSCI['Savings'].iloc[i] - MSCI['S_Invested'].iloc[i]
        
MSCI['S_Invested'].iloc[0] = 0
MSCI['S_left'].iloc[0] = 0

# Savings in UPs
MSCI['S_UPs'] = MSCI['S_left'] / MSCI['MSCI']

# Order sizing
MSCI['order'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['order'].iloc[i] = MSCI['S_UPs'].iloc[i]
    else:
        MSCI['order'].iloc[i] = None
        
MSCI['order'] = MSCI['order'].cumsum().ffill()

MSCI['order'] = MSCI['order'].fillna(0)

MSCI['UPs_new'] = MSCI['order'] + MSCI['UPs']

MSCI['Montante'] = MSCI['UPs_new'] * MSCI['MSCI']

MSCI.columns = ['Price', 'Total Savings', 'UP', 'Buy', 'Total Savings Invested', 'Total Savings Left', 'Total Savings Left in UPs', 'Order', 'UPs', 'Montante']

MSCI['Order'] = MSCI['Order'].drop_duplicates()
MSCI['Order'] = MSCI['Order'].fillna(0)

MSCI.iplot(dimensions=pa.dimensions)
```
