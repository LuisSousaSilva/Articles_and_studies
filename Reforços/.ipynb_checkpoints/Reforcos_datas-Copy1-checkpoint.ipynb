```python
%%time
starting_capital = 10000
savings = 10
start = '2014-12-31'
end   = '2020-07-31'

# Order logic
# List of days of orders
days_of_orders_list = ["2015-08-24", '2016-02-09', '2016-09-14', '2018-02-06', '2018-12-27', '2020-03-18'] #pd.date_range(start=start, end=end, freq='BM')

MSCI = ISINs = ['MSCI_World_TR']
Nomes=['MSCI']

MSCI = pd.read_csv('C:/Users/Luis/Google Drive/_GitHub/Backtester/Data/Cotacoes_diarias_all.csv', index_col='Date', parse_dates=True)[ISINs].dropna()
MSCI = Quotes[:'2020-07-31']
MSCI.columns=Nomes

MSCI = MSCI.groupby(MSCI.index).first()

# Set Period
MSCI = pa.normalize(MSCI[start:end])

# Savings
MSCI['Savings'] = savings
MSCI['Savings'].iloc[0] = 0
MSCI['Savings'] = MSCI['Savings'].cumsum()

# UPs
MSCI['UPs'] = starting_capital / MSCI['MSCI'].iloc[0]

# # Buy or hold
MSCI['Buy'] = MSCI.index.isin(days_of_orders_list)
        
# Invested
MSCI['S_Invested'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['S_Invested'].iloc[i] = MSCI['Savings'].iloc[i]
    else:
        MSCI['S_Invested'].iloc[i] = None
        
MSCI['S_Invested'] = MSCI['S_Invested'].ffill()
MSCI['S_Invested'] = MSCI['S_Invested'].shift()
MSCI['S_Invested'] = MSCI['S_Invested'].fillna(0)

# Savings left
MSCI['S_left'] = 0

for i in range(MSCI.shape[0]):
    MSCI['S_left'].iloc[i] = MSCI['Savings'].iloc[i] - MSCI['S_Invested'].iloc[i]
        
MSCI['S_Invested'].iloc[0] = 0
MSCI['S_left'].iloc[0] = 0

# Savings in UPs
MSCI['S_UPs'] = MSCI['S_left'] / MSCI['MSCI']

# Order sizing
MSCI['order'] = 0

for i in range(MSCI.shape[0]):
    if MSCI['Buy'].iloc[i]:
        MSCI['order'].iloc[i] = MSCI['S_UPs'].iloc[i]
    else:
        MSCI['order'].iloc[i] = None
        
MSCI['order'] = MSCI['order'].cumsum().ffill()
MSCI['order'] = MSCI['order'].fillna(0)
MSCI['UPs_new'] = MSCI['order'] + MSCI['UPs']
MSCI['Montante'] = MSCI['UPs_new'] * MSCI['MSCI']
MSCI['Montante Total'] = MSCI['Montante'] + MSCI['S_left']

MSCI.columns = ['Price', 'Total Savings', 'UP', 'Buy', 'Total Savings Invested', 'Total Savings Left', 'Total Savings Left in UPs', 'Order', 'UPs', 'Montante Investido', 'Montante Total']
MSCI = MSCI[['Montante Total', 'Price', 'Total Savings', 'UP', 'Buy', 'Total Savings Invested', 'Total Savings Left', 'Total Savings Left in UPs', 'Order', 'UPs', 'Montante Investido']]

MSCI['Order'] = MSCI['Order'].drop_duplicates()
MSCI['Order'] = MSCI['Order'].fillna(0)

MSCI.iplot(dimensions=pa.dimensions)
```
